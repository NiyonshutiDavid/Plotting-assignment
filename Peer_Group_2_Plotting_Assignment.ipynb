```python
# Experimenting with bin size 30
import numpy as np

co2_data = df['co2']
plt.figure(figsize=(10, 6))
plt.hist(co2_data, bins=30, edgecolor='yellow')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Number of vehicles')
plt.title('Distribution of CO2 Emissions')
plt.show
```

```python
# Experimenting with different bin sizes

# Experimenting with bin size 15

plt.figure(figsize=(10, 6))
plt.hist(co2_data, bins=15, edgecolor='black')
plt.title('Distribution of CO2 Emissions with 15 Bins')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.show()

# Experimenting with bin size 50
plt.figure(figsize=(10, 6))
plt.hist(co2_data, bins=50, edgecolor='black')
plt.title('Distribution of CO2 Emissions with 50 Bins')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.show()
```
