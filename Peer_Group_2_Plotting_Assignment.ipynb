```python
numerical_cols = df.select_dtypes(include='number')
correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap of Fuel Economy Data')
plt.show()
```

**Reasoning:**

1. df.select_dtypes(include=np.number):
This line selects only the numeric columns from the DataFrame (df). This ensures that the correlation calculation is performed only on numerical data, avoiding the error you encountered before.
2. correlation_matrix = ... .corr(): This calculates the correlation matrix for the selected numeric columns.
3.sns.heatmap(...): This part remains the same as before, creating the heatmap using the correlation matrix.



*   annot=True displays the correlation values on the heatmap
*   cmap='coolwarm' sets the color scheme to coolwarm (blue for negative correlations, red for positive correlations).


*   fmt=".2f" formats the correlation values to two decimal places.
*   Finally, we set the title and display the heatmap using plt.show().









**Interpretation of the heatmap:**
 The heatmap visualizes the correlations between different numerical variables in the fuel economy dataset.
 Key insights:
 - Strong positive correlations:
  - 'co2' (carbon dioxide emissions) and 'displ' (engine displacement): Larger engines tend to produce more CO2 emissions.
   - 'city' (city fuel efficiency) and 'highway' (highway fuel efficiency): Vehicles with good city efficiency also have good highway efficiency.
  - 'co2' and 'comb' (combined fuel efficiency): Cars with lower fuel efficiency tend to have higher CO2 emissions.
 - Strong negative correlations:
   - Fuel efficiency ('city', 'highway', 'comb') and engine displacement ('displ'): Larger engines generally have lower fuel efficiency.

# *Story told by the visualization:*
- Engine Size vs. Efficiency: Larger engines offer more power but have lower fuel efficiency and higher CO2 emissions.
 - City vs. Highway Efficiency: Improvements in engine technology or aerodynamics benefit both city and highway driving.
 - CO2 and Fuel Efficiency: CO2 emissions strongly correlate with fuel efficiency, emphasizing the environmental impact of low fuel efficiency.

**Columns to drop if the target is 'ghgScore':**
 - Since the target variable is 'ghgScore', I consider dropping the following columns:
   1. 'co2': Highly correlated with 'ghgScore'. Including it risks data leakage, leading to overly optimistic model performance.
   2. 'comb': Highly correlated with both 'co2' and 'ghgScore'. Including it can cause multicollinearity, reducing model interpretability and stability.

# *Justification:*
 - Preventing data leakage: Excluding 'co2' ensures the model predicts 'ghgScore' based on realistic features.
 - Reducing redundancy and multicollinearity: Dropping 'comb' eliminates redundancy, leading to a more robust and interpretable model.
